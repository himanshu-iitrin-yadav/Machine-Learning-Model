# What we need
### We want the line that “best fits” this data set as measured by residual sum of squares -- the simple linear regression cost. We have a closed form solution that involves the following terms:

* The number of data points (N)

* The sum (or mean) of the Ys

* The sum (or mean) of the Xs

* The sum (or mean) of the product of the Xs and the Ys

* The sum (or mean) of the Xs squared

Then once we have calculated all of these terms, we can use the formulas to compute the slope and intercept. Recall that we first solve for the slope and then we use the value of the slope to solve for the intercept.  The formula for the slope is a fraction with:

```python
N = len(x)
numerator = (((x*y).sum()) -(1/N)* ((x.sum())*(y.sum())))
denominator = (((x**2).sum()) - (1/N)*((x.sum())*(x.sum())))
```

```python
slope = numerator/denominator
slope
```

```python
intercept = y.mean() - slope*x.mean()
intercept
```

```python
plt.title("Scatter With Data And Best Fit Line")
plt.scatter(x,y , color = 'r')
plt.xlabel("x-values" , fontsize= 12)
plt.ylabel("y-values", fontsize =12)
plt.plot(x , slope*x+intercept,color='blue' , linewidth=1.8)
#plt.title("X v/s Y")
plt.grid(True)
plt.legend()
```

# gradient descent example

```python
def gradient_descent(x,y):
    n = len(x)
    current_slope = 4.8
    current_intercept = .8
    iterations = 10000
    learning_rate = 0.003
    for i in range(iterations):
        y_predicted = current_slope * x + current_intercept
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        slope_derevative = -(2/n)*sum(x*(y-(current_slope*x+current_intercept)))
        intercept_descent = -(2/n)*sum(y-(current_slope*x + current_intercept))
        current_slope = current_slope - learning_rate*slope_derevative
        current_intercept = current_intercept - learning_rate*intercept_descent
        print("m {},b {} , cost {} , iteration {}".format(current_slope,current_intercept,cost,i))
        #plt.plot(x,y_predicted)
    
```

```python
gradient_descent(x,y)
```
